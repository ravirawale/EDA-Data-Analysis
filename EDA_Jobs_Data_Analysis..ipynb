# Removing outlier from salary columns

```python
def f_per(data, va):
    ub = np.percentile(data[va], 99)
    lb = np.percentile(data[va], 1)
    return lb, ub
  
def z_score(data, va):
    ub = data[va].mean() + 2.5 * data[va].std()
    lb = data[va].mean() - 2.5 * data[va].std()
    return lb, ub
  
def iqr(data, va):
    Q1 = np.percentile(data[va], 25)
    Q3 = np.percentile(data[va], 75)
    iqr_val = Q3 - Q1
    ub = Q3 + 1.5 * iqr_val
    lb = Q1 - 1.5 * iqr_val
    return lb, ub


print("salary percentile :", f_per(df, 'salary'))
print("salary_in_usd percentile :", f_per(df, 'salary_in_usd'))
print("salary IQR :", iqr(df, 'salary'))
print("salary_in_usd IQR :", iqr(df, 'salary_in_usd'))
print("salary z_score :", z_score(df, 'salary'))
print("salary_in_usd z_score :", z_score(df, 'salary_in_usd'))
print(df['salary'].min(),df['salary_in_usd'].min())
print(df['salary'].max(),df['salary_in_usd'].max())
```

```python
df['salary']=np.clip(df['salary'],a_min=iqr(df, 'salary')[0],a_max=iqr(df, 'salary')[1])
```
